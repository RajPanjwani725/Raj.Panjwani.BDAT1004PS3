# Question 6

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

marriage_data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
```

```python
marriage_data_range=marriage_data[(marriage_data['Year']>1867) & (marriage_data['Year']<2014) ]
```

```python
marriage_data_range
```

```python
data_marriage_work=marriage_data_range[['Year','Marriages']]
data_marriage_work.set_index('Year',inplace=True)
data_marriage_work
```

```python
data_marriage_divorced=marriage_data_range[['Year','Divorces']]
data_marriage_divorced.set_index('Year',inplace=True)
data_marriage_divorced
```

```python
data_main=pd.concat([data_marriage_work,data_marriage_divorced],axis=1)
data_main.reset_index(inplace=True)
data_main
```

```python
plt.plot('Year','Marriages',data=data_main)
plt.plot('Year','Divorces',data=data_main)
plt.title("Marriage and Divorces per Year")
plt.ylabel("Count of Marriage and Divorce")
plt.xlabel("Year")
plt.legend()
plt.show()
```
